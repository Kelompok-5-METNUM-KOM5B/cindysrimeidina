```python
#Dekomposisi LU
import numpy as np 


n = int(input('Masukan Jumlah Ordo: '))
A = np.zeros((n,n+1))

x = np.zeros(n)

for i in range(n):
    for j in range(n):
        A[i][j] = float(input( 'A['+str(i)+']['+ str(j)+']='))
print('\nMatrix A : ')

# mengubah menjadi matriks 
def show(matrix):
    n = len(matrix)
    for row in range(n):
      for col in range(n):
        print('%.2f' % matrix[row][col], end="\t")
      print("")

show(A)

def decomposition(A):
  # mengetahui jumlah baris dari matriks A
  n = len(A)
  # membuat 0 
  L = [[0 for row in range(n)]
       for col in range(n)]
  U = [[0 for row in range(n)]
       for col in range(n)]
  
  for p in range(n):
    # Matriks U (segitiga atas)
    for j in range(p,n):
      # penjumlahan L(p,k)*U(k,j)
      sum = 0
      for k in range(p):
        sum = sum + L[p][k]*U[k][j]
      U[p][j] = A[p][j] - sum
    # Matriks L (segitiga bawah)
    q = p
    for i in range (q,n):
      if (i==q):
        # diagonal of L
        L[i][q]=1
      else:
        # penjumlahan L(i,k)*U(k,q)
        sum = 0
        for k in range(q):
          sum = sum + L[i][k]*U[k][q]
        L[i][q] = (A[i][q] - sum)/U[q][q]

    return L, U

L, U = decomposition(A)
# menampilkan nilai matriks
print("\nMatrix L :")
show(L)

print("\nMatrix U :")
show(U)
```

```python
#Gauss Jordan

import numpy as np
import sys


n = int(input('Masukkan Jumlah variabel: '))

a = np.zeros((n,n+1))
x = np.zeros(n)

# memasukan nilai matriks
print('Masukkan nilai:')
for i in range(n): 
    for j in range(n+1): 
        a[i][j] = float(input( 'a['+str(i)+']['+ str(j)+']='))


print ("\nMatriks persamaan:")
print (a)

#Rumus Gauss Jordan 
for i in range(n):
    if a[i][i] == 0:
        sys.exit('Terdapat nilai 0 pada diagonal utama!')
        
    for j in range(n):
        if i != j:
            ratio = a[j][i]/a[i][i]

            for k in range(n+1):
                a[j][k] = a[j][k] - ratio * a[i][k] 

for i in range(n):
    x[i] = a[i][n]/a[i][i]

print('\nSolusi SPL: ')
for i in range(n):
    print('X%d = %0.2f' %(i,x[i]), end = '\t')
```
